```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\dell\Desktop\BD_project\2018-2019_result.csv')
df
```

```python
def add_WTeam(x,y,x1,y1):
    if x1>y1:
        return x
    else:
        return y
def add_LTeam(x,y,x1,y1):
    if x1>y1:
        return y
    else:
        return x
def add_VLoc(x,y,x1,y1):
    if x1>y1:
        return 'Visitor_Team'
    else:
        return 'Home_Team'
df['WTeam']=df.apply(lambda x:add_WTeam(x.Visitor,x.Home,x.PTS,x.PTS_1),axis=1)
df['LTeam']=df.apply(lambda x:add_LTeam(x.Visitor,x.Home,x.PTS,x.PTS_1),axis=1)
df['WLoc']=df.apply(lambda x:add_VLoc(x.Visitor,x.Home,x.PTS,x.PTS_1),axis=1)
df = df.drop(['Date', 'Start','Visitor','PTS','Home','PTS_1','1','2','Attend.','Notes'], axis=1) 
df.to_csv(r'C:\Users\dell\Desktop\BD_project\2018-2019_result(new).csv') 
```
